```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
df_a = pd.read_csv("athlete_events.csv")
df_a.head()
```

```python
df_n = pd.read_csv("noc_regions.csv")
df_n.head()
```

### To join the both data frames using merge function

```python
df = df_a.merge(df_n, how ='left', on = 'NOC')
df.head()
```

### To check the rows and columns in the dataset

```python
df.shape
```

### To make the columns name consistent

```python
df.rename(columns = {'region': 'Region', 'notes': 'Notes'}, inplace=True)
df.head()
```

```python
df.describe()
```

```python
df.tail()
```

### Check the top 10 countries participating and ploting it

```python
top_10 = df.Team.value_counts().sort_values(ascending=False).head(10)
top_10
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.title('Overall Participation')
#plt.xticks(rotation = 20)
sns.barplot(x= top_10.index, y=top_10, palette = 'Set2');
```

### Age Distribution of Participants

```python
plt.figure(figsize=(12,6))
plt.xlabel('Age')
plt.ylabel('Particpants')
plt.hist(df.Age, bins = np.arange(10,80,2),color = 'orange', edgecolor='white');
```
